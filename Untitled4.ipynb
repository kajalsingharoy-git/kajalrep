```python
import pandas as pd
import networkx as nx
import itertools as it


df=pd.read_csv("C:/Users/kajal/OneDrive/Desktop/Micro Project/Output Files/Rank_n32.csv",sep=",")

df1=pd.read_csv("C:/Users/kajal/OneDrive/Desktop/Micro Project/out_n32.csv",sep=" ")
g = nx.from_pandas_edgelist(df1, 'n', 'm', edge_attr=False)


d=df["Degree"]
dr=df["Deg_Rank"]
br=df["Bw_Rank"]
cr=df["Cl_Rank"]
lr=df["Lc_Rank"]
gr=df["Gl_Rank"]

p=int(len(d)*0.1)

ndeg=dict(zip(d, dr))
nbw=dict(zip(d,br))
ncl=dict(zip(d,cr))
nlp=dict(zip(d,lr))
ngp=dict(zip(d,gr))
nodes=g.nodes()
de=g.degree()
de=set(d)




max_d=max(d)
min_d=min(d)

low_b=min_d
up_b=min_d+5

n=(max_d-low_b)/up_b

s1=s2=s3=s4=s5=c=0
davg={}
bavg={}
cavg={}
lavg={}
gavg={}

for i in de:
    if(i>=low_b & i<=up_b):
        drank=ndeg.get(i)
        s1=s1+drank
        brank=nbw.get(i)
        s2=s2+brank
        crank=ncl.get(i)
        s3=s3+crank
        lrank=nlp.get(i)
        s4=s4+lrank
        grank=ngp.get(i)
        s5=s5+grank
        c=c+1
    avg1=s1/c
    davg[i]=avg1
    avg2=s2/c
    bavg[i]=avg2
    avg3=s3/c
    cavg[i]=avg3
    avg4=s4/c
    lavg[i]=avg4
    avg5=s5/c
    gavg[i]=avg5
    s1=s2=s3=s4=s5=c=0
    low_b=low_b+5
    up_b=up_b+5
    
    
adeg=list(davg.values())
abw=list(bavg.values())
acl=list(cavg.values())
alp=list(lavg.values())
agp=list(gavg.values())

de=list(de)
de1=de
de=["Degree"]+de
adeg=["Avg Rank of Deg_Rank"]+adeg
abw=["Avg Rank of Bw_Rank"]+abw
acl=["Avg Rank of Cl_Rank"]+acl
alp=["Avg Rank of Lc_Rank"]+alp
agp=["Avg Rank of Gl_Rank"]+agp


de1.sort()
new_d={}

i=0
sum_of_rank=0

while (i<max_d):
    for i in de1:
        if(i>=low_b & i<=up_b):
            cur_rank=davg[i]
            sum_of_rank=sum_of_rank+cur_rank
            c=c+1
        avg_rank=int(sum_of_rank/c)
        print("rank of ",i,"is: ",avg_rank)
    new_d[up_b]=avg_rank
    low_b=up_b+1
    up_b=low_b+p

print("\n\n", new_d,"\n\n")    
```
