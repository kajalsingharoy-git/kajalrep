```python
import pandas as pd
import networkx as nx
import itertools as it


df=pd.read_csv("C:/Users/kajal/OneDrive/Desktop/Micro Project/out_n32.csv",sep=" ")
g = nx.from_pandas_edgelist(df, 'n', 'm', edge_attr=False)

g1=g

nx.draw_circular(g1, with_labels=True)

nodes=g.nodes()
nodes=list(nodes)
nodes.sort()
nodes2=nodes

l=len(nodes2)

nodes_list=["Nodes"]
for n in nodes:
    nodes_list.append(n)

    
#CALCULATE DEGREE OF ALL NODES---------------------------------------------------------------------------------------
 

deg=sorted(g.degree, key=lambda x: x[1], reverse=True)


deg=dict(deg)

deg_list=[]

for i in deg.keys():
    deg_list.append(i) #create list of nodes sorted according to no. of degrees in descending order
    

    
#CALCULATE CLOSENESS CENTRALITY---------------------------------------------------------------------------------------


path_length = nx.single_source_shortest_path_length #creating a macro for inbuilt function

cl_cent = {} 

for n in nodes: 
    sp = path_length(g,n) #calculating distance of each connected node from current node
    
    totsp = sum(sp.values()) #calculate total distance of all nodes concted to current node
    
    if totsp > 0.0 and len(g) > 1: 
        cl_cent[n] = (len(sp)-1.0) / totsp #calculate closeness centrality
    else: 
        cl_cent[n] = 0.0
        
cl_cent = sorted(cl_cent.items(), key=lambda kv: kv[1], reverse=True)

cl_cent=dict(cl_cent)

cl_list=[]

for i in cl_cent.keys():
    cl_list.append(i)
    

#CALCULATE BETWEENNES CENTRALITY-------------------------------------------------------------------------------------


        

node_and_times = dict.fromkeys(g.nodes(), 0) 
sum_paths = 0

for s, t in it.product(g.nodes(), repeat=2):
    paths = nx.all_shortest_paths(g, s, t) 
    
    for path in paths:
        sum_paths += 1
        
        
        for node in path[1:-1]: 
            node_and_times[node] += 1
        
bw_cent = {k : v*1.0/sum_paths for k, v in node_and_times.items()}

   
bw_cent = sorted(bw_cent.items(), key=lambda kv: kv[1], reverse=True) 

bw_cent=dict(bw_cent)

bw_list=[]

for i in bw_cent.keys():
    bw_list.append(i)
    


#CALCULATE LOCAL PROPERTIES OF NODES---------------------------------------------------------------------------------



d1=g.degree()
d2=dict()
d3=dict()
d4=dict()
  

for n in nodes:
    net=g.neighbors(n)
    net=list(net)
    net
    d2[n]=d1[n]
    
    for j in net:
        d2[n]=d2[n]+g.degree(j)
        
for n in nodes:
    net=g.neighbors(n)
    net=list(net)
    net
    d3[n]=d2[n]
    
    for j in net:
        d3[n]=d3[n]+d2[j]
        
for n in nodes:
    net=g.neighbors(n)
    net=list(net)
    net
    d4[n]=d3[n]
    
    for j in net:
        d4[n]=d4[n]+d3[j]
        
lc_prop=sorted(d4.items(), key=lambda kv: kv[1],reverse=True)

lc_prop=dict(lc_prop)

lc_list=[]

for i in lc_prop.keys():
    lc_list.append(i)



#CALCULATE GLOBAL PROPERTIES OF NODES---------------------------------------------------------------------------------


    
md=min(deg.values()) #check min degree value

node_removed=[]



noe=g.number_of_edges()
l=len(nodes)

while l>0:
    
        for n in nodes: 
            if g.degree(n)==md: #checking whether n has min degree
                g.remove_node(n)
                node_removed.append(n) #create list of removed nodes               
            noe=g.number_of_edges()
            if l>0: #recalculating min degree value
                deg=g.degree()
                deg=dict(deg)
                deg=dict(sorted(deg.items()))
                md=min(deg.values())
            l=l-1
nodes=g.nodes()

for n in nodes:
    node_removed.append(n) #adding last remaining node into list to complete it.
    
node_removed.reverse() #reversing to calculate rank from 1 to end

gl_list=[]

for n in node_removed:
    gl_list.append(n) #create a standardized named list for easier recognition


#PRINT TO FILE ACCORDING TO RANK--------------------------------------------------------------------------------------

curr_deg=dict(d1)
neighbour_deg=dict(d2)
curr_deg=list(curr_deg.values())
neighbour_deg=list(neighbour_deg.values())

curr_deg.insert(0,"Degree")
neighbour_deg.insert(0,"Neighbours' Degree")
deg_rank=["Deg_Rank"]
cl_rank=["Cl_Rank"]
bw_rank=["Bw_Rank"]
lc_rank=["Lc_Rank"]
gl_rank=["Gl_Rank"]


for n in nodes2: 
    rank_of_curr_node=deg_list.index(n)+1
    deg_rank.append(rank_of_curr_node)
        
    rank_of_curr_node=cl_list.index(n)+1
    cl_rank.append(rank_of_curr_node)
        
    rank_of_curr_node=bw_list.index(n)+1
    bw_rank.append(rank_of_curr_node)
    
    rank_of_curr_node=lc_list.index(n)+1
    lc_rank.append(rank_of_curr_node)
    
    rank_of_curr_node=node_removed.index(n)+1
    gl_rank.append(rank_of_curr_node)
    

    
print("Degree Rank:", deg_rank, "\n\n")
print("Closeness Rank:", cl_rank, "\n\n")
print("Betweenness Rank:", bw_rank, "\n\n")
print("Local Rank:", lc_rank, "\n\n")
print("Global Rank:", gl_rank, "\n\n") 

output = '\n'.join('\t'.join(map(str,row)) for row in zip(nodes_list,curr_deg,neighbour_deg,deg_rank,bw_rank,cl_rank,lc_rank,gl_rank))

with open('Rank_n32.csv', 'w') as f:
    f.write(output)
     
curr_deg=list(curr_deg)


print(output)
```
