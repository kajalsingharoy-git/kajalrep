```python
import pandas as pd
import networkx as nx
import random as rd


df=pd.read_csv(r"C:\Users\kajal\OneDrive\Desktop\Project\e\adoh.csv",sep=" ") #read csv
g = nx.from_pandas_edgelist(df, 'n', 'm', edge_attr=False) #make network from csv

nodes=g.nodes() #get list of nodes
a=dict(g.degree()) #make dictionary of nodes:degree
b=list(a.values()) #make list of degrees
c=list(a.values()) #make another list of degrees

b.sort(reverse=True) #sort degrees in desc order

maxdeg=b[0] #get max degree

rbnodes={}

for i in nodes:
    rbnodes[i]=round((c[i-1]/maxdeg)*100) # make dictionary of node:robustness value
        
infected = [] #infected list
jlist=[1] #list of infected nodes whose neighbours are to be calculated 
ch = [] 

for i in range(1,16): # Day ranging from day1 to day5
    print ("\n Day " + str(i) + " started!\n")
    print("length of jlist", len(jlist))
    for j in jlist: # to search the infected neighbours
        infected.append(j)
        neigh=list(g.neighbors(j)) # taking neighbours of j
        #print("\nNeighbours of",j,": ", neigh)
        #print("length of neighbours: ", len(neigh))
        for k in neigh: # k is taking neighbours of j 
            rn=rd.randint(1,101) # taking random integer
            dd=rbnodes.get(k) # get robustness value of k node
            #print("\nrandom no: ",rn)
            #print("\nrobustness of node",k,": ",dd)
            if rn>=dd: # comparing the random integer number with robustness of the k node
                ch.append(k)
            ch = list(set(ch)) # taking unique set of ch values
        #print ("Infected Neighbours of :" + str(j))
        #print (ch)
          
    for ele in ch: # elements in ch list
        #print (ele)
        if ele in infected: # element is infected
            ch.remove(ele) # remove prior infected nodes
    jlist = (ch) # add new list of infected nodes
    #print ("jlist: ")
    #print (jlist) # list of infected nodes
    #print("length of jlist: ", len(jlist))
    #print("infected nodes of", i, ":", infected)
    print("length of infected: ", len(infected))
    ch = [] #Emptying ch list
    
```
