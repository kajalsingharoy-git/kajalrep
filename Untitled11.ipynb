```python
import pandas as pd
import networkx as nx
import random as rd


df=pd.read_csv(r"C:\Users\kajal\OneDrive\Desktop\MAJOR_PROJECT\NEW\UC.csv",sep=" ") #read csv
g = nx.from_pandas_edgelist(df, 'n', 'm', edge_attr=False)#make network from csv

nodes=g.nodes() #get list of nodes
a=dict(g.degree()) #make dictionary of nodes:degree
b=list(a.values()) #make list of degrees
c=list(a.values()) #make another list of degrees

b.sort(reverse=True) #sort degrees in desc order

maxdeg=b[0] #get max degree

rbnodes={}

for i in nodes:
    rbnodes[i]=50 #get robustness of each node w.r.t. degree
        
        

infected = {} #make list of infected nodes
immune=[] #make list of immune nodes
jlist=[1] #make another list of infected nodes
ch = [] #make list of unique infected nodes
setinf=[]

for i in range(1,9):
    print ("\n\nDay " + str(i) + " started!\n")
    for j in jlist:
        if j not in immune:
            infected[j]=5
            neigh=list(g.neighbors(j))
            for k in neigh:
                rn=rd.randint(1,101)
                dd=rbnodes.get(k)
                if rn>=dd and k not in immune:
                    ch.append(k)
                ch = list(set(ch))
                
    #print("infected", infected)
    #print("immune", immune)
    print("length of infected", len(infected))
    print("length of immune", len(immune))
    
    for k in infected:
        temp=infected.get(k)
        temp=temp-1
        infected[k]=temp
        
    
    for ele in list(infected):
        if infected.get(ele)<1:
            infected.pop(ele)
            immune.append(ele)
            
    for ele in ch:
        if ele in infected:
            ch.remove(ele)
    jlist = (ch)
    ch = []   
```
